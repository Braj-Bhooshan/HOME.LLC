```python
def extract_year(df):
    df['date'] = pd.to_datetime(df['date'], format=('%Y-%m-%d'))
    df['year'] = df['date'].dt.year
    df_column = list(df.columns)
    df1 = df.groupby('year').mean().reset_index()
    df = df1[df_column[-2:][::-1]]

    return df
```

```python
df_house= extract_year(df_house)
df_price_index = extract_year(df_price_index)
df_income = extract_year(df_income)
df_per_saving = extract_year(df_per_saving)
df_population = extract_year(df_population)
df_uneply = extract_year(df_uneply)
```

```python
df = pd.merge(df_house,df_income,on='year',how='outer')
df= pd.merge(df,df_per_saving,on='year',how='outer')
df = pd.merge(df,df_population, on='year',how='outer')
df = pd.merge(df,df_uneply, on ='year',how='outer')
df = pd.merge(df,df_em_pop,on='year',how='outer')
df = pd.merge(df,df_price_index,on='year',how='outer')
```

```python
df.sort_values('year',inplace=True)
df.reset_index(drop=True,inplace=True)
```
